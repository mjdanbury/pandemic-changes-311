```python
for up in went_up:
    df = pd.read_csv('../data/'+str(up)+'.csv',parse_dates=[0])
    df['avg'] = df['count'].rolling(window=30).mean()
    chart = alt.Chart(df).mark_line(clip=True).encode(
        x=alt.X('day',axis=alt.Axis(grid=False,tickCount=3,title="")).scale(domain=(pd.to_datetime('2016'),pd.to_datetime('2024'))),
        y=alt.Y('avg',axis=alt.Axis(grid=True,title="",domain=False,tickOpacity=0))
    ).properties(title=df.iloc[0]['complaint_type'] + ': ' + df.iloc[0]['descriptor'],width=200,height=150).configure_view(
        stroke=None
    )
    chart.save('../charts-up/'+str(up)+'.svg')
```

```python
for down in went_down:
    df = pd.read_csv('../data/'+str(down)+'.csv',parse_dates=[0])
    df['avg'] = df['count'].rolling(window=30).mean()
    chart = alt.Chart(df).mark_line(clip=True).encode(
        x=alt.X('day',axis=alt.Axis(grid=False,tickCount=3,title="")).scale(domain=(pd.to_datetime('2016'),pd.to_datetime('2024'))),
        y=alt.Y('avg',axis=alt.Axis(grid=True,title="",domain=False,tickOpacity=0))
    ).properties(title=df.iloc[0]['complaint_type'] + ': ' + df.iloc[0]['descriptor'],width=200,height=150).configure_view(
        stroke=None
    )
    chart.save('../charts-down/'+str(down)+'.svg')
```
